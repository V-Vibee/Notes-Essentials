### slicing

```python
sales.loc[:, ['Sales', 'Profit']]
```

```python
# Getting only head

sales.loc[:, ['Sales', 'Profit']].head()
```

```python
sales.iloc[:, [0,3,2]].head()
```

```python
#picking up the rows or slicing up the rows

sales.loc[['Western Africa','Southern Africa', 'North Africa'], :]
```

```python
sales.iloc[0:3, :]
```

```python
# slicing both columns and rows i.e 2D

sales.loc[['Western Africa','Southern Africa', 'North Africa'], ["Sales","Profit"]]
```

```python
sales.iloc[0:3,2:4]
```
